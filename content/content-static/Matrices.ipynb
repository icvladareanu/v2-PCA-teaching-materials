# 2.2. Matrices and Transformations

## What is a linear transformation?

A **transformation** (or function) $T$ takes an input vector $\mathbf{x}$ and assigns it to an output vector $T(\mathbf{x})$.

A transformation is **linear** if it satisfies two algebraic rules for all vectors $\mathbf{u}, \mathbf{v}$ and any scalar $c$:

1.  **Preserves addition:** $T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})$
2.  **Preserves scalar multiplication:** $T(c\mathbf{u}) = cT(\mathbf{u})$

```{tip} Geometrical intuition 
:class: dropdown
:icon: true
:open: true

**Geometrically**, a linear transformation essentially moves an input vector to an output vector. You can imagine all of two-dimensional space as an infinite grid, where each vector represents an arrow, or a dot (coordinate set).
To understand what a specific transformation does, you have to understand how it affects every possible vector in space simultaneously.
A transformation is linear if it keeps the "grid lines of space" parallel and evenly spaced, and leaves the origin $(0,0)$ fixed in place. It essentially keeps the space flat without curving it.
```



## What is a matrix transformation?

A matrix transformation is a specific type of linear transformation that can be represented using matrix multiplication.
It turns out that *every* linear transformation from an $n$-dimensional space to an $m$-dimensional space can be perfectly described by an $m \times n$ matrix $A$, such that:

$$
T(\mathbf{x}) = A\mathbf{x}
$$

This means every input vector x is transformed by multiplying it with a matrix A.

---

**Geometrically**, to understand what a specific matrix does to a dataset, you only need to ask one question: **Where do the basis vectors go?**

```{tip} What are basis vectors again?
:class: dropdown
:icon: true
:open: false

In a standard 2D Cartesian coordinate system, every vector can be described as a combination of two special unit vectors, called the **basis vectors**:
*   $\hat{i}$ (i-hat): the x-axis vector $\begin{bmatrix} 1 \\ 0 \end{bmatrix}$
*   $\hat{j}$ (j-hat): the y-axis vector $\begin{bmatrix} 0 \\ 1 \end{bmatrix}$
```
Any vector can be written as a linear combination of these basis vectors. After the transformation, the vector lands at the **same linear combination of the transformed basis vectors**.

If

$$
\mathbf{v} = -1\hat{i} + 2\hat{j}
$$

then after the transformation:

$$
T(\mathbf{v}) = -1\,T(\hat{i}) + 2\,T(\hat{j})
$$

So by knowing where $\hat{i}$ and $\hat{j}$ go, you can determine where **any vector goes**.

## Eigenvectors and Eigenvalues

(eigenvectors)=
An *eigenvector* of a matrix is a non-zero vector whose direction does not change in a geometric sense when the matrix is applied to it. After the transformation, the vector remains on the same line (its span is preserved), but it may be stretched, shrunk, or flipped in the opposite direction. 

Most vectors do not have this property: when a linear transformation is applied, they are typically rotated and moved off the line they originally span. Eigenvectors are special because they are exactly the directions that are not rotated by the transformation. 

(eigenvalues)=
An *eigenvalue* is a scalar that describes how much an eigenvector is stretched or compressed during a linear transformation. A negative eigenvalue means the vector is also flipped while being scaled. 

In the equation:

$$
\mathbf{A}\mathbf{v} = \lambda \mathbf{v}
$$

- $\mathbf{v}$ is the eigenvector of matrix $\mathbf{A}$
- $\lambda$ is the eigenvalue corresponding to that eigenvector $\mathbf{v}$

---


```{admonition} Example: Computing eigenvalues and eigenvectors
:class: dropdown
:open: false
:icon: true

Consider the matrix:

$$
\mathbf{A} =
\begin{bmatrix}
3 & 1 \\
1 & 3
\end{bmatrix}
$$

---

### Step 1: Eigenvalues

$$
\det(\mathbf{A} - \lambda \mathbf{I}) = 0
$$

$$
\det \begin{bmatrix} 3 - \lambda & 1 \\ 1 & 3 - \lambda \end{bmatrix} = 0
$$

$$
(3 - \lambda)(3 - \lambda) - (1)(1) = 0
$$

$$
\lambda^2 - 6\lambda + 8 = 0 \Rightarrow (\lambda - 4)(\lambda - 2) = 0
$$

$$
\lambda_1 = 4, \quad \lambda_2 = 2
$$

---

### Step 2: Eigenvectors

We solve:

$$
(\mathbf{A} - \lambda \mathbf{I})
\begin{bmatrix}
v_1 \\
v_2
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

---

#### For $\lambda = 4$

$$
\begin{bmatrix} 3 - 4 & 1 \\ 1 & 3 - 4 \end{bmatrix} \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} = \begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$

$$
-v_1 + v_2 = 0 \Rightarrow v_2 = v_1
$$

$$
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

---

#### For $\lambda = 2$

$$
\begin{bmatrix} 3 - 2 & 1 \\ 1 & 3 - 2 \end{bmatrix} \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} = \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$

$$
v_1 + v_2 = 0 \Rightarrow v_2 = -v_1
$$

$$
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
$$
